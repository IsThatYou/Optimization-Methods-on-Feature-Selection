## 5 Genre classification

### 5.1 From features

```python
small = tracks['set', 'subset'] <= 'small'

train = tracks['set', 'split'] == 'training'
val = tracks['set', 'split'] == 'validation'
test = tracks['set', 'split'] == 'test'

y_train = tracks.loc[small & train, ('track', 'genre_top')]
y_test = tracks.loc[small & test, ('track', 'genre_top')]
enc = skl.preprocessing.LabelEncoder()
y_train = enc.fit_transform(y_train)
y_test = enc.transform(y_test)

X_train = features.loc[small & train,]
X_test = features.loc[small & test,]

print('{} training examples, {} testing examples'.format(y_train.size, y_test.size))
print('{} features, {} classes'.format(X_train.shape[1], np.unique(y_train).size))
```

```python
# Be sure training samples are shuffled.
X_train, y_train = skl.utils.shuffle(X_train, y_train, random_state=42)

# Standardize features by removing the mean and scaling to unit variance.
scaler = skl.preprocessing.StandardScaler(copy=False)
scaler.fit_transform(X_train)
scaler.transform(X_test)

# Support vector classification.
clf = skl.svm.SVC()
clf.fit(X_train, y_train)
score = clf.score(X_test, y_test)
print('Accuracy: {:.2%}'.format(score))

# rfcc = RandomForestClassifier(max_depth=2, random_state=0)
# rfcc.fit(X_train, y_train)
# rfc_score = rfcc.score(X_test, y_test)
# print('Accuracy: {:.2%}'.format(rfc_score))

# from sklearn.tree import DecisionTreeClassifier
# clf = DecisionTreeClassifier(random_state=0)
# clf.fit(X_train, y_train)
# score = clf.score(X_test, y_test)
# print('Accuracy: {:.2%}'.format(score))

# from sklearn.neural_network import MLPClassifier
# nn = MLPClassifier(solver='adam', alpha=1e-5,hidden_layer_sizes=(2000, 30), random_state=1)
# nn.fit(X_train,y_train)
# score = nn.score(X_test, y_test)
# print('Accuracy: {:.2%}'.format(score))
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LassoCV
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestClassifier
# from sklearn.linear_model import RandomizedLasso
```

```python
rfc = RandomForestClassifier(max_depth=2, random_state=0)
rfc.fit(X_train,y_train)
importances = rfc.feature_importances_
indices = np.argsort(importances)[::-1]
plt.figure(dpi = 520)
plt.title("Feature importances (FMA)")

plt.bar(range(X_train.shape[1]), importances[indices],
       color="r", align="center")
plt.xticks(range(X_train.shape[1]), indices)
plt.xlim([-1, 23])
plt.xlabel("index of the feature")
plt.savefig('foo.png')
plt.show()

####try something other than svm

alpha_ridge = [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20]
lasso = LassoCV(alphas = alpha_ridge,max_iter=1000)
elastic_net = ElasticNet(alpha=0.01,l1_ratio = 0.5)
# lasso2 = RandomizedLasso(alpha=0.025)

sfm = SelectFromModel(lasso, threshold="1*mean")
sfm.fit(X_train, y_train)
n_features = sfm.transform(X_train).shape[1]
# print(rfc.feature_importances_)
print(n_features)
X_transform = sfm.transform(X_train)

# print(X_transform)
```

```python
# clf1 = RandomForestClassifier(max_depth=2, random_state=0)
clf1 = skl.svm.SVC()
clf1.fit(X_transform, y_train)
score = clf1.score(sfm.transform(X_test), y_test)
print('Accuracy: {:.2%}'.format(score))
```
